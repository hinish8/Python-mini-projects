## TF-IDF 

```python
df2.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#max_features=20000
tfidf = TfidfVectorizer(stop_words="english", max_features=10000)
df2["overview"] = df2["overview"].fillna(" ")

tfidf_matrix = tfidf.fit_transform(df2["overview"])
tfidf_matrix.shape
```

```python
tfidf.get_feature_names_out()[-10:]
```

```python
tfidf.vocabulary_ 
```

```python
# Cosine similarity calculation
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
cosine_sim
```

```python
# Used later to retrieve movie names based on the index
movie2idx = pd.Series(df2.index, index=df2["title"]).drop_duplicates()
```

```python
movie2idx.head(10)
```

```python
# Function to Return Top 10 Most Similar Movies
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movie2idx[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df2["title"].iloc[movie_indices]
```

```python

get_recommendations("The Intern")
```

```python

get_recommendations("The Devil Wears Prada")
```
